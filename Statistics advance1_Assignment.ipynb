13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

 --> The Central Limit Theorem (CLT) is one of the most fundamental concepts in statistics and probability theory. It states that:

  When we take a sufficiently large number of random samples from any population (regardless of its original distribution), the   sampling distribution of the sample mean will be approximately normal.

  This means that even if the original population is not normally distributed, the distribution of sample means will become       approximately normal as the sample size increases.
  ### Why is CLT Important?
  
  
  1. Foundation for Statistical Inference:
  
      Many statistical methods (confidence intervals, hypothesis testing, regression analysis, etc.) rely on the assumption of         normality.
     
      CLT allows us to use normal distribution approximations, even when the underlying data is not normally distributed.
       
  
  2. Justification for Using the Normal Distribution:
  
       Many real-world datasets are not normally distributed (e.g., income, stock prices, waiting times).
   
       CLT enables us to apply normal-based methods to these non-normal data by using sample means.
       
   
  3. Practical Applications in Real Life:
  
     Polling & Surveys:

            When estimating the average opinion of a population, the sample mean follows a normal distribution.
            
     Quality Control:

            Manufacturers use CLT to monitor production processes and ensure products meet specifications.
  
     Finance & Economics:

            Stock market returns, risk modeling, and economic indicators often rely on sample averages, making CLT 
            essential in financial modeling.
            
  The Central Limit Theorem is critical because it allows us to analyze data from any population distribution using normal         distribution techniques, provided we have a sufficiently large sample size.

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_z_scores(data):
    """Calculate Z-scores for a given dataset."""
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Using sample standard deviation (ddof=1)
    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(z_scores):
    """Plot histogram of Z-scores."""
    plt.hist(z_scores, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black', label='Histogram')

    # Overlay standard normal distribution curve
    x = np.linspace(-4, 4, 1000)
    pdf = stats.norm.pdf(x, 0, 1)  # Standard normal distribution (mean=0, std=1)
    plt.plot(x, pdf, 'r-', lw=2, label='Standard Normal PDF')

    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.title('Z-score Distribution')
    plt.legend()
    plt.grid(True)
    plt.show()

# Generate a sample dataset
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=1000)  # Normal distribution with mean=50, std=10

# Calculate Z-scores and plot them
z_scores = calculate_z_scores(sample_data)
plot_z_scores(z_scores)
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def central_limit_theorem_demo(sample_size, num_samples):
    """Demonstrate the Central Limit Theorem (CLT) using a uniform distribution."""
    
    # Generate a uniform distribution (non-normal parent distribution)
    population = np.random.uniform(0, 10, 100000)  # Large uniform population

    # Collect sample means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))

    # Plot histogram of sample means
    sns.histplot(sample_means, bins=30, kde=True, color='blue', edgecolor='black')
    plt.axvline(np.mean(sample_means), color='red', linestyle='dashed', linewidth=2, label='Mean')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.title(f'CLT Demonstration (Sample Size={sample_size}, Num Samples={num_samples})')
    plt.legend()
    plt.grid(True)
    plt.show()

# Parameters
sample_size = 30  # Number of observations per sample
num_samples = 1000  # Number of samples

# Run CLT demonstration
central_limit_theorem_demo(sample_size, num_samples)
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test.
    
    Parameters:
    sample : list or numpy array -> Sample dataset
    population_mean : float -> Known population mean (H0)
    population_std : float -> Known population standard deviation
    alpha : float -> Significance level (default 0.05)
    
    Returns:
    Z-score, p-value, and test decision (reject or fail to reject H0)
    """
    
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    
    # Compute Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    
    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    
    # Decision based on alpha
    decision = "Reject H0" if p_value < alpha else "Fail to Reject H0"
    
    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.3f}")
    print(f"P-value: {p_value:.5f}")
    print(f"Decision: {decision}")

    return z_score, p_value, decision

# Example usage
sample_data = np.array([65, 68, 70, 72, 62, 75, 60, 71, 69, 73])  # Sample dataset
population_mean = 70   # Hypothesized population mean (H0)
population_std = 5     # Known population standard deviation
alpha = 0.05          # Significance level

# Perform hypothesis testing
z_test(sample_data, population_mean, population_std, alpha)
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95, population_std=None):
    """
    Compute the confidence interval for a dataset.
    
    Parameters:
    - data: list or numpy array -> Sample dataset
    - confidence: float -> Confidence level (default: 0.95)
    - population_std: float (optional) -> Known population standard deviation (use Z-distribution)
    
    Returns:
    - Lower and upper bounds of the confidence interval
    """
    sample_mean = np.mean(data)
    sample_size = len(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)

    # Choose Z or t distribution based on knowledge of population standard deviation
    if population_std:
        standard_error = population_std / np.sqrt(sample_size)
        critical_value = stats.norm.ppf((1 + confidence) / 2)  # Z-critical value
    else:
        standard_error = sample_std / np.sqrt(sample_size)
        critical_value = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)  # t-critical value
    
    # Compute margin of error
    margin_of_error = critical_value * standard_error

    # Compute confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Confidence Level: {confidence*100:.1f}%")
    print(f"Margin of Error: {margin_of_error:.2f}")
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return lower_bound, upper_bound

# Example dataset
sample_data = np.array([65, 68, 70, 72, 62, 75, 60, 71, 69, 73])  # Sample dataset
confidence_level = 0.95  # 95% confidence level

# Calculate confidence interval (assuming unknown population std)
confidence_interval(sample_data, confidence_level)
```

```python
#17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # For reproducibility
mean_true = 50       # True population mean
std_true = 10        # True population standard deviation
sample_size = 30     # Sample size

sample_data = np.random.normal(mean_true, std_true, sample_size)

# Step 2: Compute the confidence interval
def confidence_interval(data, confidence=0.95):
    """
    Compute the confidence interval for a normally distributed dataset.

    Parameters:
    - data: list or numpy array -> Sample dataset
    - confidence: float -> Confidence level (default: 0.95)

    Returns:
    - Lower and upper bounds of the confidence interval
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (unbiased)
    sample_size = len(data)

    # Use t-distribution for small sample sizes (n < 30)
    critical_value = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)

    # Compute margin of error
    standard_error = sample_std / np.sqrt(sample_size)
    margin_of_error = critical_value * standard_error

    # Compute confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Confidence Level: {confidence*100:.1f}%")
    print(f"Margin of Error: {margin_of_error:.2f}")
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return lower_bound, upper_bound

# Compute 95% confidence interval
ci_lower, ci_upper = confidence_interval(sample_data)

# Step 3: Visualize the data and confidence interval
plt.hist(sample_data, bins=10, alpha=0.7, color='blue', edgecolor='black', density=True)
plt.axvline(np.mean(sample_data), color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label='Lower CI Bound')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label='Upper CI Bound')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Sample Data with Confidence Interval')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Define the normal distribution parameters
mean = 0     # Mean (Î¼)
std_dev = 1  # Standard deviation (Ïƒ)

# Step 2: Generate x values (range of possible values)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)  # Covering ~99.99% of values

# Step 3: Compute the Probability Density Function (PDF)
pdf_values = stats.norm.pdf(x, mean, std_dev)

# Step 4: Plot the PDF
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_values, color='blue', label=f'Normal Distribution\nMean={mean}, Std={std_dev}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of Normal Distribution')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Define the Poisson distribution parameter (Î» = expected number of events)
lambda_poisson = 5  # Mean (Î») of Poisson distribution

# Step 2: Generate x values (possible number of occurrences)
x_values = np.arange(0, 15)  # Consider values from 0 to 14

# Step 3: Compute the Cumulative Distribution Function (CDF)
cdf_values = stats.poisson.cdf(x_values, lambda_poisson)

# Step 4: Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x_values, cdf_values, where='mid', color='blue', label=f'Poisson CDF (Î»={lambda_poisson})')
plt.xlabel('Number of Events (x)')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) of Poisson Distribution')
plt.axhline(y=0.5, color='red', linestyle='dashed', linewidth=1, label='50% Probability Line')
plt.xticks(x_values)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Step 5: Interpretation of CDF values
for x in [3, 5, 7]:
    prob = stats.poisson.cdf(x, lambda_poisson)
    print(f"P(X â‰¤ {x}) = {prob:.4f} -> Probability that events â‰¤ {x}")
```

```python
#20. Simulate a random variable using a continuous uniform distribution and calculate its expected value. 

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the continuous uniform distribution parameters
a, b = 2, 10  # Lower bound (a) and Upper bound (b)

# Step 2: Generate random samples from Uniform(a, b)
n_samples = 10000  # Number of samples
samples = np.random.uniform(a, b, n_samples)

# Step 3: Compute the Expected Value (Theoretical and Empirical)
expected_value_theoretical = (a + b) / 2  # Formula: E[X] = (a + b) / 2
expected_value_empirical = np.mean(samples)  # Empirical mean from generated samples

# Step 4: Plot the histogram of the uniform distribution
plt.figure(figsize=(8, 5))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Continuous Uniform Distribution')
plt.axvline(expected_value_empirical, color='red', linestyle='dashed', linewidth=2, label='Empirical Mean')
plt.axvline(expected_value_theoretical, color='green', linestyle='dashed', linewidth=2, label='Theoretical Mean')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Step 5: Print the Expected Values
print(f"Theoretical Expected Value: {expected_value_theoretical:.4f}")
print(f"Empirical Expected Value (from simulation): {expected_value_empirical:.4f}")
```

```python
#21. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate two datasets with different standard deviations
np.random.seed(42)  # For reproducibility

# Dataset 1: Lower standard deviation (Ïƒ = 5)
mean1, std_dev1, size1 = 50, 5, 1000
data1 = np.random.normal(mean1, std_dev1, size1)

# Dataset 2: Higher standard deviation (Ïƒ = 15)
mean2, std_dev2, size2 = 50, 15, 1000
data2 = np.random.normal(mean2, std_dev2, size2)

# Step 2: Compute standard deviations
std_dev1_empirical = np.std(data1, ddof=1)  # Sample standard deviation
std_dev2_empirical = np.std(data2, ddof=1)

# Step 3: Plot histograms to compare distributions
plt.figure(figsize=(8, 5))
plt.hist(data1, bins=30, alpha=0.6, color='blue', edgecolor='black', label=f'Dataset 1 (Ïƒ â‰ˆ {std_dev1_empirical:.2f})')
plt.hist(data2, bins=30, alpha=0.6, color='red', edgecolor='black', label=f'Dataset 2 (Ïƒ â‰ˆ {std_dev2_empirical:.2f})')

# Step 4: Customize plot
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Comparison of Two Datasets with Different Standard Deviations')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Step 5: Print Standard Deviations
print(f"Standard Deviation of Dataset 1: {std_dev1_empirical:.4f}")
print(f"Standard Deviation of Dataset 2: {std_dev2_empirical:.4f}")
```

```python
#22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a normal distribution dataset
np.random.seed(42)  # For reproducibility
mean, std_dev, size = 50, 10, 1000  # Mean=50, Std Dev=10, Sample size=1000
data = np.random.normal(mean, std_dev, size)

# Step 2: Calculate Range and Interquartile Range (IQR)
data_range = np.max(data) - np.min(data)  # Range = max - min
q1 = np.percentile(data, 25)  # 25th percentile (Q1)
q3 = np.percentile(data, 75)  # 75th percentile (Q3)
iqr = q3 - q1  # Interquartile Range (IQR)

# Step 3: Plot the distribution with boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(data, color="blue")
plt.title("Boxplot of Normally Distributed Data")
plt.xlabel("Values")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Step 4: Print the Results
print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
```

```python
#23. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a normal distribution dataset (Before normalization)
np.random.seed(42)  # For reproducibility
mean, std_dev, size = 50, 15, 1000  # Mean=50, Std Dev=15, Sample size=1000
data = np.random.normal(mean, std_dev, size)

# Step 2: Apply Z-score Normalization
mean_data = np.mean(data)
std_data = np.std(data, ddof=1)  # Sample standard deviation
normalized_data = (data - mean_data) / std_data  # Z-score transformation

# Step 3: Plot the original and normalized distributions
plt.figure(figsize=(10, 5))

# Original Data
plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color='blue', edgecolor='black')
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Normalized Data
plt.subplot(1, 2, 2)
sns.histplot(normalized_data, bins=30, kde=True, color='green', edgecolor='black')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Step 4: Print the Statistics
print(f"Original Data: Mean = {mean_data:.2f}, Std Dev = {std_data:.2f}")
print(f"Normalized Data: Mean = {np.mean(normalized_data):.2f}, Std Dev = {np.std(normalized_data, ddof=1):.2f}")
```

```python
#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

def generate_normal_data(mean=50, std_dev=10, size=1000):
    """Generate a dataset from a normal distribution."""
    np.random.seed(42)  # For reproducibility
    return np.random.normal(mean, std_dev, size)

def calculate_skewness_kurtosis(data):
    """Calculate and return skewness and kurtosis of the dataset."""
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # Excess kurtosis (subtracts 3)
    return data_skewness, data_kurtosis

def plot_distribution(data):
    """Plot histogram and KDE of the dataset."""
    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, kde=True, color="blue", edgecolor="black")
    plt.title("Histogram of Normally Distributed Data")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

# Generate data
data = generate_normal_data()

# Calculate skewness and kurtosis
skewness, kurt = calculate_skewness_kurtosis(data)

# Plot the distribution
plot_distribution(data)

# Print results
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis (Excess): {kurt:.4f}")
```
